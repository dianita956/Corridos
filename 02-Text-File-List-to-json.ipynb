```python
import json
#file to convert
file="List of unincorporated communities in Texas.txt" 
#create a list to store data
dictionary= {}
unincorparatedcommunitieslist = []
#read file
with open(file) as f:
    for d in f:
        #print(d)
        #read lines from file, creating vars based on position and indicator of info needed
        parts = d.strip().split("|", 2)

        if len(parts)<3:
            name = parts[0] if len(parts)>0 else ' '
            county = parts[1] if len(parts) >1 else ''
            other = ''
        else:
            name, county, other = parts
       
        x={
            "Community Name":name,
            "County": county
        }
        unincorparatedcommunitieslist.append(x)

dictionary["unincorparated communities"] = unincorparatedcommunitieslist

print("number of communities:", len(dictionary["unincorparated communities"]))
#create json output file
out_file = open("../Corridos/data/TexasInuncorparatedCommunities.json", "w")
json.dump(dictionary, out_file, indent=4) 
out_file.close()

#run code
```
